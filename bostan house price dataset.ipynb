```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
bostanData = pd.read_csv("train.csv")
```

```python
bostanData.head()
```

```python
x = bostanData.drop(['SalePrice'],axis=1)
y = bostanData[['SalePrice']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,random_state = 1212121)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
y_train
```

# processing of training Data Sets

```python
bostanDataCategrical = x_train.select_dtypes(object)
bostanDataNumerical = x_train.select_dtypes(np.number)
```

```python
bostanDataCategrical.isna().sum()
```

```python
bostanDataCategrical.drop(['PoolQC','Fence','MiscFeature',
                           'Alley'],axis=1,inplace=True)
```

# imputation of numerical columns

```python
from sklearn.impute import SimpleImputer
simputer = SimpleImputer(strategy="most_frequent")
```

```python
simputer.fit(bostanDataCategrical)
```

```python
tree = simputer.transform(bostanDataCategrical)
bostanDataCategricalNullRemove = pd.DataFrame(tree,columns=bostanDataCategrical.columns)
```

```python
bostanDataCategricalNullRemove.isnull().sum()
```

# Imputation of Numerical Columns

```python
simputer1 = SimpleImputer(strategy="mean")
```

```python
simputer1.fit(bostanDataNumerical)
```

```python
tree1 = simputer1.transform(bostanDataNumerical)
bostanDataNumericalNullRemovel = pd.DataFrame(tree1,columns=bostanDataNumerical.columns)
```

```python
bostanDataNumericalNullRemovel.isnull().sum()
```

# Data Processing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
bostanDataCategricalEncoded = bostanDataCategricalNullRemove.apply(le.fit_transform)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
ss.fit(bostanDataNumericalNullRemovel)
```

```python
let = ss.transform(bostanDataNumericalNullRemovel)
bostanDataNumericalEncoder = pd.DataFrame(let,columns=bostanDataNumericalNullRemovel.columns)
```

```python
bostanDataNumericalEncoder
```

```python
bostanDataFinel = pd.concat([bostanDataCategricalEncoded,bostanDataNumericalEncoder],axis=1)
```

```python
bostanDataFinel.isna().sum()
```

```python
bostanDataFinel
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression,SGDClassifier
lr = LinearRegression()
sgd = SGDClassifier()
```

```python
lr.fit(bostanDataFinel.to_numpy(),y_train.to_numpy())
```

```python
sgd.fit(bostanDataFinel.to_numpy(),y_train.to_numpy())
```

# Imputation of Numerical Columns

```python
tree1 = simputer1.transform(bostanDataNumericalTest)
bostanDataNumericalNullRemoveltest = pd.DataFrame(tree1,columns=bostanDataNumericalTest.columns)
```

```python
bostanDataNumericalNullRemoveltest.isnull().sum()
```

# Data Processing

```python
bostanDataCategricalEncoded = bostanDataCategricalNullRemovetest.apply(le.fit_transform)
```

```python
let = ss.transform(bostanDataNumericalNullRemoveltest)
bostanDataNumericalEncoder = pd.DataFrame(let,columns=bostanDataNumericalNullRemoveltest.columns)
```

```python
bostanDataNumericalEncoder
```

```python
bostanDataFinel1 = pd.concat([bostanDataCategricalEncoded,bostanDataNumericalEncoder],axis=1)
```

```python
prediction_lr = lr.predict(bostanDataFinel1)
prediction_sgd = sgd.predict(bostanDataFinel1)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
r2_score(prediction_lr,y_test)
```

```python
np.sqrt(mean_squared_error(prediction_lr,y_test))
```

```python
np.sqrt(mean_squared_error(prediction_sgd,y_test))
```
